```python
import scipy.stats as stats
```

```python
n = 200  # number of trials
p = 0.02  # probability of success (i.e., selecting a noise image X-ray)
k = 2  # minimum number of successes we want to calculate the probability for
```

```python
# Calculate the probability of selecting two or more noise image X-rays
prob = 1 - stats.binom.cdf(k-1, n, p)
```

```python
print(f"The probability of selecting two or more noise image X-rays is {prob:.4f}")
```
